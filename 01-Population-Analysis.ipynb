```python
import pandas as pd
import matplotlib.pyplot as plt

# IMPORTANT: skiprows=4 is required
df = pd.read_csv(
    r"API_SP.POP.TOTL_DS2_en_csv_v2_61.csv",
    skiprows=4
)

print("Dataset Loaded Successfully!")
# Show first rows
print(df.head())
year = '2020'
df_year = df[['Country Name', year]]

# Remove missing values
df_year = df_year.dropna()
print("\nData after cleaning:")
print(df_year.head())

top10 = df_year.sort_values(
    by=year,
    ascending=False
).head(10)

# BAR CHART — Top 10 Countries
plt.figure(figsize=(12,6))
plt.bar(
    top10['Country Name'],
    top10[year]
)

plt.title("Top 10 Countries by Population (2020)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# HISTOGRAM — Population Distribution
plt.figure(figsize=(8,5))
plt.hist(
    df_year[year],
    bins=20
)

plt.title("Population Distribution Histogram (2020)")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
